```python
import xlrd
# comentarios
#path: direccion de archivo

import os
import pandas as pd
import numpy as np
import array as arr
import altair as alt

pd.set_option("display.max_columns",None)
```

```python
gastoIES = pd.read_excel("n2017_PY_Gasto_en_Salud_por_Fuente_de_Financiamiento_1993-2017.xlsx")
gastoIES=gastoIES[["Anio","Estado_Abreviado","EstadoPrint","IES_GPTS_porcentaje_PIB","PIBmxn","IES_GPTS_porcentaje_GPT","IES_GPS_PERCAPITA_Total_n2017"]]
gastoIES=gastoIES[gastoIES.Estado_Abreviado != "EUM"].rename(columns={"Estado_Abreviado":"Estado","IES_GPTS_porcentaje_PIB":"PIB","IES_GPTS_porcentaje_GPT":"GPT","IES_GPS_PERCAPITA_Total_n2017":"PERCAPITA"})
```

```python
def map_values(value):
    if value == "BC" or value == "BCS" or value == "CHIH" or value == "DGO" or value == "SIN" or value == "SON" :
        return 'Noroeste'
    elif value == "COAH" or value == "NL" or value == "TAMPS":
        return 'Noreste'
    elif value == "COL" or value == "JAL" or value == "MICH" or value == "NAY":
        return 'Oeste'
    elif value == "HGO" or value == "PUE" or value == "TLAX" or value == "VER":
        return 'Oriente'
    elif value == "AGS" or value == "GTO" or value == "QRO" or value == "SLP" or value == "ZAC":
        return 'Centronorte'
    elif value == "CDMX" or value == "EDOMEX" or value == "MOR":
        return 'Centrosur'
    elif value == "CHIS" or value == "GRO" or value == "OAX":
        return 'Suroeste'
    elif value == "CAMP" or value == "QROO" or value == "TAB" or value == "YUC":
        return 'Sureste'
    else:
        return 'Unknown'
    
```

```python
gastoIES["Region"]=gastoIES["Estado"].apply(map_values)
```

```python
gastoIES
```

```python
periodo = 6
sexenios = [i for i in range(gastoIES.Anio.min(), gastoIES.Anio.max()+periodo, periodo)]
sexenios=[1995,2001,2007,2013,2018]
print(sexenios)

sexeniosCol = pd.cut(gastoIES.Anio, bins=sexenios, precision=0, right=False)
gastoIES.insert(1, "sexenio", sexeniosCol)
gastoIES.head()
```
