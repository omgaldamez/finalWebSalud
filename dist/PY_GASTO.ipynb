```python
import xlrd
# comentarios
#path: direccion de archivo

import os
import pandas as pd
import numpy as np
import array as arr
import altair as alt

pd.set_option("display.max_columns",None)
```

```python
gastoIES = pd.read_excel("n2017_PY_Gasto_en_Salud_por_Fuente_de_Financiamiento_1993-2017.xlsx")
```

```python
gastoIES=gastoIES[["Anio","Estado_Abreviado","IES_GPTS_porcentaje_PIB","IES_GPTS_porcentaje_GPT","IES_GPS_PERCAPITA_Total_n2017"]]
gastoIES=gastoIES[gastoIES.Estado_Abreviado != "EUM"].rename(columns={"Estado_Abreviado":"Estado","IES_GPTS_porcentaje_PIB":"PIB","IES_GPTS_porcentaje_GPT":"GPT","IES_GPS_PERCAPITA_Total_n2017":"PERCAPITA"}).set_index("Estado")
```

```python

gastoIES
```

```python

gastoIES.to_json(r"JSON_GASTO.json",orient="table")
```

```python
with pd.ExcelWriter("GASTO_CSV.xlsx") as writer:
  gastoIES.to_excel(writer, sheet_name="gasto")
```

```python
gastoIES[["PIB","Anio"]]
```

```python
gastoIES
```

```python
gastoPERC = pd.DataFrame(gastoIES.groupby("Estado")["PERCAPITA"].mean())
gastoPERC = gastoPERC.sort_values(by="PERCAPITA",ascending=True)
gastoPERC["Orden"]=gastoPERC.reset_index().index+1
gastoPERC=pd.merge(gastoIES,gastoPERC,on="Estado",how="inner").sort_values(by=["Orden","Anio"],ascending=True)
gastoPERC.rename(columns={"PERCAPITA_x":"PERCAPITA","PERCAPITA_y":"PERCAPITA_mean"},inplace=True)
gastoPERC
```

```python
gastoPERC.to_json(r"JSON_PERC.json",orient="table")
```

```python
gastoPIB = pd.DataFrame(gastoIES.groupby("Estado")["PIB"].mean())
gastoPIB = gastoPIB.sort_values(by="PIB",ascending=True)
gastoPIB["Orden"]=gastoPIB.reset_index().index+1
gastoPIB=pd.merge(gastoIES,gastoPIB,on="Estado",how="inner").sort_values(by=["Orden","Anio"],ascending=True)
gastoPIB.rename(columns={"PIB_x":"PIB","PIB_y":"PIB_mean"},inplace=True)
gastoPIB
```

```python
gastoPIB.to_json(r"JSON_PIB.json",orient="table")
```

```python
gastoGPT = pd.DataFrame(gastoIES.groupby("Estado")["GPT"].mean())
gastoGPT = gastoGPT.sort_values(by="GPT",ascending=True)
gastoGPT["Orden"]=gastoGPT.reset_index().index+1
gastoGPT=pd.merge(gastoIES,gastoGPT,on="Estado",how="inner").sort_values(by=["Orden","Anio"],ascending=True)
gastoGPT.rename(columns={"GPT_x":"GPT","GPT_y":"GPT_mean"},inplace=True)
gastoGPT
```

```python
gastoGPT.to_json(r"JSON_GPT.json",orient="table")
```

```python
gastoALF = pd.DataFrame(gastoIES.groupby("Estado")["GPT"].mean())
gastoALF = gastoALF.sort_values(by="Estado",ascending=True)
gastoALF["Orden"]=gastoALF.reset_index().index+1
gastoALF=pd.merge(gastoIES,gastoALF,on="Estado",how="inner").sort_values(by=["Estado","Anio"],ascending=True)
gastoALF.rename(columns={"GPT_x":"GPT","GPT_y":"GPT_mean"},inplace=True)
gastoALF
```

```python
gastoALF.to_json(r"JSON_ALF.json",orient="table")
```

```python
with pd.ExcelWriter("GASTO_CSV.xlsx") as writer:
  gastoPERC.to_excel(writer, sheet_name="PERC")
  gastoPIB.to_excel(writer, sheet_name="PIB")
  gastoGPT.to_excel(writer, sheet_name="GPT")
  gastoALF.to_excel(writer, sheet_name="ALF")
```

```python
periodo = 6
sexenios = [i for i in range(gastoIES.Anio.min(), gastoIES.Anio.max()+periodo, periodo)]
sexenios=[1995,2001,2007,2013,2018]
print(sexenios)

sexeniosCol = pd.cut(gastoIES.Anio, bins=sexenios, precision=0, right=False)
gastoIES.insert(1, "sexenio", sexeniosCol)
gastoIES.head()
```

```python
gastoIES
```

```python
presis = [
    "Ernesto Zedillo Ponce de León", 
    "Vicente Fox Quesada", 
    "Felipe Calderón Hinojosa",
    "Enrique Peña Nieto"]

presidenteCol = pd.cut(gastoIES.Anio, bins=sexenios, precision=0, right=False, labels=presis)
gastoIES.insert(1, "presidente", presidenteCol)
gastoIES.head()
Arrow=gastoIES
```
